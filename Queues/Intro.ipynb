```python
from collections import deque
# list is not recommended for queue - because in some cases list can be slow

q = deque()

#enqueue - add to back of queue
q.append(10)
q.append(20)
q.append(30)

print(q)
```

```python
#front - either use appendleft or use index

q.appendleft(0)
print(q)
```

```python
# deque - remove from front

q.popleft()
print(q)
```

```python
# check empty
print(len(q) == 0)
```
