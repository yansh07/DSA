### The search space is halved at each step.

**But only on one condition:**  
1. The array/search space must be sorted (or monotonic in nature).

### MOST IMPORTANT RULE
1. sorted array
2. minimum / maximum possible
3. first / last occurrence
4. smallest value such that condition is true
5. monotonic
6. search space

Time Complexity - O(log n)

````python
def binary_search(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

print(binary_search([0, 1, 2, 3, 4, 5], 5))
````
